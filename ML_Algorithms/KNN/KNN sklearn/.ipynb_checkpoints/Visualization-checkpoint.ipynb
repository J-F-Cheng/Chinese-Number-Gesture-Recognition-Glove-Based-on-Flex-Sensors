```python
fig = plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(121)
ax1.title.set_text('Data Collected by Junfeng Cheng')
plt.ylim(250,750)

newall['thumb'] = newall['thumb'] - 70
newall['index'] = newall['index'] + 60
newall['middle'] = newall['middle'] + 60
newall['ring'] = newall['ring'] + 30
newall['little'] = newall['little'] + 20

sns.boxplot(data=fingerdataall.drop(columns="gesture"),order=["thumb","index","middle","ring","little"])
ax2 = fig.add_subplot(122)
sns.boxplot(data=newall.drop(columns="gesture"))
ax2.title.set_text('Data Collected by Jiatai Zhang, after calibration')
plt.ylim(250,750)
```
